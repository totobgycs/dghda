```python
import tensorflow as tf
import pandas as pd
import matplotlib as plt
import urllib.request, json
```

What data we use:
  * biggest stocks from EUROSTOX50, NASDAQ and DOW (to limit input dimensionality)
  * for the last 5 years (to have growing and volaile)
  * daily oclh
Where do we get them:
  AlphaVantage



1. Get json data for Ahold-Delhaize (symbol: AD.AMS) from alphavantage
2. Putting into a pandas datafame (just relevant data)
3. Rename columns, and
4. Transform numeric values to numeric.

```python
with urllib.request.urlopen(
       "https://www.alphavantage.co/query?"
           "function=TIME_SERIES_DAILY&"
           "symbol=AD.AMS&"
           "outputsize=full&"
           "apikey=SLQLZPOJI7PGOISP") as url:
    data = json.loads(url.read().decode())
    df=pd.DataFrame.from_dict(data['Time Series (Daily)'], orient='index')
    df.rename(index=str, columns={'1. open': 'op', '2. high': 'hi', '3. low': 'lo', '4. close': 'cl', '5. volume': 'vl'}, inplace=True)
    for s in df.columns:
        df[s] = pd.to_numeric(df[s])
```
